```python
group_data = data.groupby('Summary').mean().sort_values('Temperature (C)')
mean_temp_by_group = group_data['Temperature (C)']
mean_temp_by_group.plot(kind='bar')
```

```python
column_name = 'Temperature (C)'
data[column_name].hist(bins=20)
```

```python
data[column_name].plot(kind= 'density')
```

```python
column_name = 'Humidity'
skewness = skew(data[column_name])
kurt = kurtosis(data[column_name])

data[column_name].plot(kind='density')
print(f'Skewness = {skewness}. Kurtosis = {kurt}')
```

```python
col1 = 'Temperature (C)'
col2 = 'Humidity'

corr = np.corrcoef(data[col1], data[col2])
print(corr)
data.plot(kind='scatter', x=col1, y=col2)
```
